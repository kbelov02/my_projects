## Задание №1  
Переведите содержимое файла purchase_log.txt* в словарь purchases вида:  
{‘1840e0b9d4’: ‘Продукты’, …}  
Учтите условия на данные:  
- содержимое purchase_log.txt* помещается в оперативную память компьютера

```python
import json

# пустой список, в который будем записывать данные из файла
list_from_purchase=[]
i = 0
with open('purchase_log.txt') as f:
# проходим циклом по каждой строчке файла
  for line in f:
# удаляем перенос строки и лишние проблеы
    line = line.strip()
    dict_ = json.loads(line)
    for value in dict_.values():
# заносим значения в список
      list_from_purchase.append(value)
    i += 1
```

```python
# проверим, что получилось: 
list_from_purchase[:6]
```

```python
# преобразуем список в словарь и удаляем из него пару ключ-значение, в которой содержится названия пар словаря: 
res_dict = {list_from_purchase[i]: list_from_purchase[i+1] for i in range(0, len(list_from_purchase), 2)}
del(res_dict['user_id'])
```

```python
# для проверки результата выводим первые 10 пар нашего словаря: 
counter = 0
for i, k in res_dict.items():
  print(i, k)
  counter += 1
  if counter == 10:
    break
```

## Задание №2  
Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки, если покупка была, сам файл visit_log.csv* изменять не надо.   
Запишите в файл funnel.csv визиты из файла visit_log.csv*, в которых были покупки с указанием категории.  
Учтите условия на данные:  
содержимое visit_log.csv* — не помещаентся в оперативную память; используйте только построчную обработку этого файла.

```python
# открываем файл с визитами на чтение
with open('visit_log.csv', 'r') as visit_log:
# создаем файл с воронкой на запись
  with open('funnel.csv', 'w') as funnel:
    for line in visit_log:
# разделяем записи в файле с визитами по запятой и удалем лишние пробелы
      line = line.strip().split(',')
# проверка на покупку пользователем и создание колонки с категорией покупки
      if line[0] in res_dict.keys(): 
        line.append(res_dict[line[0]])
# запись получившихся данных в файл с воронкой
        funnel.write(str(line) + '\n')
    funnel.close()     
```
