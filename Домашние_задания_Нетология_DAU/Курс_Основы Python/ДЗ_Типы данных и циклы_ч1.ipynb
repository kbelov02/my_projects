## Задание 1
Дана переменная, в которой хранится слово из латинских букв. Напишите код,который выводит на экран:  
среднюю букву, если число букв в слове нечетное;  
две средних буквы, если число букв четное.

```python
word = input('Введите слово, состоящее из латинских букв ')

if word.isalpha() and len(word) >0:
  if len(word) % 2 == 0:
    print(word[len(word) // 2 - 1: len(word) // 2 +1 ])
  else:
    print(word[len(word) //2])
else:
  print('Некорректный ввод слова')
print('Конец программы')
```

## Задание 2
Напишите программу, которая последовательно запрашивает у пользователя числа (по одному за раз) и после первого нуля выводит сумму всех ранее введенных чисел.

```python
num = int(input('Введите целое число '))

while num != 0:
  num1 = int(input('Введите целое число '))
  num = num + num1
  if num1 == 0:
    break
print(num)
print('Конец программы')
```

## Задание 3
Мы делаем MVP dating-сервиса, и у нас есть список парней и девушек.  
Выдвигаем гипотезу: лучшие рекомендации мы получим, если просто отсортируем имена по алфавиту и познакомим людей с одинаковыми индексами после сортировки!  
Но мы не будем никого знакомить, если кто-то может остаться без пары:

```python
girls = list(map(str, input('Введите имена девушек через пробел ').split()))
boys = list(map(str, input('Введите имена парней через пробел ').split()))

if len(girls) !=0 and len(boys) !=0:
  girls = sorted(girls)
  boys = sorted(boys)
  if len(girls) == len(boys):
    print('Идеальные пары:')
    ind_girls = 0
    ind_boys = 0
    while ind_girls <= (len(girls) -1):
       print(girls[ind_girls], 'и', boys[ind_boys])
       ind_girls += 1
       ind_boys += 1
  else:
    print('Внимание, кто-то может остаться без пары!')
else:
  print('Некорректный ввод данных')
print('Конец программы')
```

## Задание 5 (необязательное)
Имеется список с транспортными номерами. Необходимо написать код, который проверит каждый номер на валидность (1 буква, 3 цифры, 2 буквы, 2-3 цифры).  
Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.  
Если номер валиден, то вывести его в нужном формавте (пример ниже), а если не валиден — вывести текст. При решении помогут регулярные выражения, с которыми будет знакомство на практике.

```python
from types import NoneType
import re

car_ids = ['А222ВС190', 'А22ВС190', 'м504аВ799', 'У907Кт13', '38-70ОГО', 'А15ВВ90']
x = 0
while x <=(len(car_ids)-1):
  # проверка на количество символов в номере. он может состоять как из восьми: А000АА99, так и из 9 символов: А000АА199. В последнем случае символ региона не будет начинаться с нуля. 
  if 8 < len(car_ids[x]) <= 9: 
    # проверка на соответствие элемента списка регулярному выражению. Выражение разбиваем на 2 группы, чтобы в дальнейшем реализовать отдельно вывод региона
    result = re.match(r'([авекмнорстухАВЕКМНОРСТУХ]{1}\d{3}[авекмнорстухАВЕКМНОРСТУХ]{2})([1-9]\d{1,2})', car_ids[x])
    # если элемент не прошел проверку, то тип перменной result будет не определен, что будет являться опказателем не валидности номера
    if isinstance(result, NoneType):
      print('Номер', car_ids[x], 'не валиден')
      x += 1
    else:
      # для эстетики все буквы в выводе переводим в верхний регистр: на вход нам могут быть поданы номера, с буквами, написанными в нижнем регистре. Валидными такие номера от этого они быть не перестанут.  
      print('Номер ', result.group(1).upper(), 'валиден. Регион: ', result.group(2))
      x += 1
  else:
    print('Номер', car_ids[x], 'не валиден')
    x += 1
print('Конец программы')
```

## Задание 6 (необязательное)
Дан поток логов по количеству просмотренных страниц для каждого пользователя (пользователь,дата;просмотры).  
Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей.  
И тут регулярные выражения облегчат немного реализацию.

```python
import re
stream = [ 'user4,2021-01-01;3', 'user3,2022-01-07;4', 'user2,2022-03-29;1', 'user1,2020-04-04;13', 'user2,2022-01-05;7', 'user1,2021-06-14;4', 'user3,2022-07-02;10', 'user4,2021-03-21;19', 'user4,2022-03-22;4', 'user4,2022-04-22;8', 'user4,2021-05-03;9', 'user4,2022-05-11;11' ] 
#stream = [ 'user100,2022-01-01;150', 'user99,2022-01-07;205', 'user1001,2022-03-29;81' ]  - второй лог для проверки
new_list = []
# создаем двухуровневый список, каждый элемент которого - данные о пользователе, дате и количестве просмотров
for element in stream:
  result = re.split(r',|;', element)
  new_list.append(result)
# создаем два новых списка: первый - с пользователями, второй с просмотрами
list_of_users = []
list_of_shows = []
for element in new_list:
  list_of_users.append(element[0])
  list_of_shows.append(element[2])

# создаем список с уникальными пользователями
unique_list_of_users = list(set(list_of_users))

# преобразуем элементы списка с просмотрами в целые числа, чтобы можно было совершать математические операции
list_of_shows = list(map(int, list_of_shows))
#выводим финальный результат
print('Среднее количество просмотров на уникального пользователя:', round(sum(list_of_shows) / len(unique_list_of_users),2))
print('Конец программы')
```
