# Домашнее задание к лекции «Функции»

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

Перечень всех документов.  
documents = [
{‘type’: ‘passport’, ‘number’: ‘2207 876234’, ‘name’: ‘Василий Гупкин’},  
{‘type’: ‘invoice’, ‘number’: ‘11-2’, ‘name’: ‘Геннадий Покемонов’},  
{‘type’: ‘insurance’, ‘number’: ‘10006’, ‘name’: ‘Аристарх Павлов’}  
] 

Перечень полок, на которых хранятся документы. Если документ есть в documents, то он обязательно должен быть и в directories.  
directories = {  
‘1’: [‘2207 876234’, ‘11-2’],  
‘2’: [‘10006’],  
‘3’: []  
}  
Общие требования к программе:  
* код должен быть грамотно декомпозирован: каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется;
* в коде отсутствуют глобальные переменные, за исключением documents и directories;
* пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу вводом команды «q».

## Задание 1

```python
# объявляем массивы данных
documents = [ {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'}, {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'}, {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}]
directories = {'1': ['2207 876234', '11-2'], '2': ['10006'], '3': []}
```

```python
#объявляем стартовую функцию
def chose_comand():
    """
    Данная функция служит для ввода команд и вызова функций работы системы
    """
    comand = 0
    while comand !='q':
      comand = input('Введите команду: ')
      if comand == 'q':
        result = print('Спасибо за то что пользовались моей программой. До новых встреч!')
      elif comand == 'p':
        search_doc_owner(documents)
      elif comand == 's':
        search_shelf_number(directories)
      elif comand == 'l':
        doc_information(documents, directories)
      elif comand == 'ads':
        shelf_add(directories)
      elif comand == 'ds':
        shelf_terminator(directories)
      elif comand == 'ad':
        document_adding(documents, directories)
      elif comand == 'd':
        file_eraser(documents, directories)
      elif comand == 'm':
        replacer(directories)
      else: 
        result = print('Неправильная команда')
    return result
```

### Пункт 1. Пользователь по команде «p» может узнать владельца документа по его номеру.

```python
def search_doc_owner(docs):
    doc_num = input('Введите номер документа ')
    for record in docs:
      if record['number'] == doc_num:
        result = print('Владелец документа: ', record['name'])
        break
    else:
      result = print('Документ не найден в базе')
    return result 
```

```python
chose_comand()
```

### Пункт 2. Пользователь по команде «s» может по номеру документа узнать, на какой полке он хранится.

```python
def search_shelf_number(dir):
    doc_num = input('Введите номер документа ')
    for shelf, docs in dir.items():
        for num in docs:
          if num == doc_num:
            result = print('Документ хранится на полке ', shelf)
            break 
        else:
          result = print('Документ не найден в базе')
        return result
```

```python
chose_comand()
```

### Пункт 3. Пользователь по команде «l» может увидеть полную информацию по всем документам.

```python
def doc_information(docs, dir):
    for i in docs:
      for k in dir.items():
        if i['number'] in k[1]:
          result = print(f"№: {i['number']}, тип: {i['type']}, владелец: {i['name']}, полка хранения: {k[0]}")
    return result
```

```python
chose_comand()
```

### Пункт 4. Пользователь по команде «ads» может добавить новую полку.

```python
def shelf_add(shelves):
    new_shelf_num = input('Введите номер полки: ')
    if new_shelf_num not in shelves.keys():
      shelves[new_shelf_num] = []
      result = print(f'Текущая полка добавлена. Новый перечень полок: {", ".join(list(shelves.keys()))}')
    else: 
      result = print(f'Такая полка уже существует. Текущий перечень полок: {", ".join(list(shelves.keys()))}')
    return result
```

```python
chose_comand()
```

### Пункт 5. Пользователь по команде «ds» может удалить существующую полку из данных, только если она пустая.

```python
def shelf_terminator(shelves):
    del_shelf_num = input('Введите номер полки, которую хотите удалить ')
    if del_shelf_num in shelves.keys():
      if not shelves[del_shelf_num]:
        del shelves[del_shelf_num]
        result = print(f'Полка удалена. Текущий перечень полок: {", ".join(list(shelves.keys()))}')
      else:
        result = print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {", ".join(list(shelves.keys()))}')
    else:
      result = print(f'Такой полки не существует. Текущий перечень полок: {", ".join(list(shelves.keys()))}')
    return result
```

```python
chose_comand()
```

### Пункт 1. Пользователь по команде «ad» может добавить новый документ в данные.

```python
documents = [ {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'}, {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'}, {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}]
directories = {'1': ['2207 876234', '11-2'], '2': ['10006'], '3': []}
```

```python
def document_adding(docs, shelves):
# проверка существования документа
    doc_num = input('Введите номер документа ')
    flag = 0
    for record in docs:
      if record['number'] == doc_num:
        for key, values in shelves.items():
          for value in values:
            if value == doc_num:
              flag = 1
              result = print('Такой документ уже есть в базе. Полка хранения', key)
              print('Текущие документы в базе: ')
              doc_information(documents, directories)
              break
    # проверка существования полки
    if flag == 0:
      shelf_num = input('Введите полку для хранения ')
      if shelf_num not in shelves.keys():
        result = print('Такой полки не существует. Для добавления полки используйте команду "ads"')
        print('Текущие документы в базе: ')
        doc_information(documents, directories)
      else: 
        flag = 1
    # добавляем документ в базу
        if flag == 1:
          doc_type = input('Введите тип документа ')
          doc_owner = input('Введите владельца документа ')
          docs.append({'type': doc_type, 'number': doc_num, 'name': doc_owner.title()})
          shelves[shelf_num].append(doc_num)
          result = print(f'Документ за номером {doc_num} успешно добавлен на полку {shelf_num}')
          print('Текущие документы в базе: ')
          doc_information(documents, directories)
    return result
```
