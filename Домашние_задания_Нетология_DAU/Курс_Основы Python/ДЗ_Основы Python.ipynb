## Задание 1 
Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2.
Напишите код, который проверяет какая из этих строк длиннее.

```python
phrase_1 = input('Введите фразу 1 ')
phrase_2 = input('Введите фразу 2 ')

if len(phrase_1) > len(phrase_2):
  print('Фраза 1 длиннее фразы 2')
elif len(phrase_1) < len(phrase_2):
  print('Фраза 2 длиннее фразы 1')
else:
  print('Фразы имеют одинаковую длину')
print('Конец программы')
```

## Задание 4
Вам нужно написать программу для подбора упаковок по размерам товара. Размеры (ширина, длина, высота) хранятся в переменных (в сантиметрах):

Используйте следующие правила:

* если каждое из трех измерений менее или равно 15 сантиметрам, то выведите на  экран "Коробка №1";
* если **хотя бы одно** из измерений больше 2 метров, то выводите "Упаковка для лыж";
* если **хотя бы одно** из измерений больше 15 сантиметров, но менее 50 сантиметров, то выводите "Коробка №2";
* во всех остальных случаях выводите "Коробка №3".


```python
width = int(input('Введите значение ширины, см '))
length = int(input('Введите значение длины, см '))
height = int(input('введите значение высоты, см '))

if width <= 0 or length <=0 or height <=0:
  print('Введите корректные значения')
else:
  if width <= 15 and length <=15 and height <=15:
    print('Коробка №1')
  elif width > 200 or length > 200 or height > 200:
    print('Упаковка для лыж')
  elif 15 < width > 50 or 15 < length < 50 or 15 < height < 50:
    print('Коробка №2')
  else:
    print('Коробка №3')
print('Конец программы')
```

## Задание 5 (необязательное)  
### Вариант №1: решение с использованием целочисленного деления и деления с отстатком
Дана переменная, в которой хранится шестизначное число (номер проездного билета).  
Напишите программу, которая будет определять, является ли данный билет "счастливым".  
Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.

```python
ticket = input('Введите шестизначный номер билета ')

if ticket.isdigit() and len(ticket) == 6:
  num1 = int(ticket) // 100000
  num2 = int(ticket) // 10000 % 10
  num3 = int(ticket) // 1000 % 10
  num4 = int(ticket) // 100 % 10
  num5 = int(ticket) // 10 % 10
  num6 = int(ticket) % 10
  if num1 + num2 + num3 == num4 + num5 + num6:
    print('Вам выпал счастливый билет')
  else:
    print('Вам выпал обычный билет')
else:
  print('Некорректный ввод данных')
print('Конец программы')
```

### Вариант №2: решение с использованием индексов
Дана переменная, в которой хранится шестизначное число (номер проездного билета).  
Напишите программу, которая будет определять, является ли данный билет "счастливым".  
 Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.


```python
ticket = input('Введите шестизначный номер билета ')

if ticket.isdigit() and len(ticket) == 6:
  if int(ticket[0]) + int(ticket[1]) + int(ticket[2]) == int(ticket[3]) + int(ticket[4]) + int(ticket[5]):
    print('Вам выпал счастливый билет')
  else:
    print('Вам выпал обычный билет')
else:
  print('Некорректный ввод данных') 
print('Конец программы')
```

## Задание 6 (необязательное)
Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

если пользователь выбрал круг, запрашиваем его радиус,  
если треугольник – длины трех его сторон,  
если прямоугольник – длины двух его сторон.


```python
figure = input('Введите название фигуры (круг, треугольник или прямоугольник) ')

if figure in ['Треугольник', 'треугольник']:
  side_a = input('Введите длину стороны a ')
  side_b = input('Введите длину стороны b ')
  side_c = input('Введите длину стороны c ')
  if side_a.isdigit() and side_b.isdigit() and side_c.isdigit():
   side_a = float(side_a)
   side_b = float(side_b)
   side_c = float(side_c)
   if (side_a + side_b > side_c) and (side_a + side_c > side_b) and (side_b + side_c > side_a) and side_a > 0 and side_a > 0 and side_c > 0:
    half_perim = (side_a+side_b+side_c)/2
    print('Площадь равна ', (half_perim*(half_perim-side_a)*(half_perim-side_b)*(half_perim-side_c))**0.5)
   else:
      print('Треугольника с такими сторонами не существует')
  else:
    print('Треугольника с такими сторонами не существует')
elif figure in ['Круг', 'круг']:
  radius = input('Введите радиус ')
  if radius.isdigit(): 
   radius = float(radius) 
   if radius > 0:
    print('Площадь равна ', 3.141592*(radius**2))
   else:
    print('Круга с таким радиусом не существует')
  else:
    print('Круга с таким радиусом не существует')
elif figure in ['Прямоугольник', 'прямоугольник']:
  side_x = input('Введите длину стороны x ')
  side_y = input('Введите длину стороны y ')
  if side_x.isdigit() and side_y.isdigit():
   side_x=float(side_x) 
   side_y=float(side_y)
   if side_x > 0 and side_y > 0:
    print('Площадь равна ', side_x*side_y)
   else:
    print('Прямоугольника с такими стороными не существует')
  else:
    print('Прямоугольника с такими стороными не существует')
else:
  print('Неправильно задана фигура')
print('Конец программы')
```
