## Задание 2
Дана переменная, в которой хранится список поисковых запросов пользователя (пример структуры данных приведен ниже, но запросы потенциально могут быть любые).  
Вам необходимо написать программу, которая выведет на экран распределение количества слов в запросах в требуемом виде.

```python
import re
queries = [ 'смотреть сериалы онлайн', 'новости спорта', 'афиша кино', 'курс доллара', 'сериалы этим летом', 'курс по питону', 'сериалы про спорт', 'мама мыла раму, а папа читал газету', ] 

new_list = []
for i in queries:
  result = re.split(r' ', i)
  new_list.append(result)

list_of_len = []
for i in new_list:
  list_of_len.append(len(i))

dict_ = {}

for i in list_of_len:
  if i in dict_:
    dict_[i] +=1
  else:
    dict_[i] = 1
for key, value in sorted(dict_.items()):
  print('Поисковых запросов, содержащих ', key, ' слов(а)', round((value / sum(dict_.values()))*100, 2),'%')

```

```python
my_list = ['2018-01-01', 'yandex', 'cpc', 100]
my_list1 =['a', 'b', 'c', 'd', 'e', 'f']
```

```python
# создаем словарь, ключ которого - предпоследний элемент списка, а значение - последний
tree_dict = {my_list[-2]:my_list[-1]}

# в цикле перебираем список от пред-предпоследнего элемента к первому (перебор в обратном порядке) и на каждом шаге создаем новый ключ
for i in my_list[:-2][::-1]:
  tree_dict = {i:tree_dict}
print(tree_dict)
```

## Задание 6 (необязательно)
Дана книга рецептов с информацией о том, сколько ингредиентов нужно для приготовления блюда в расчете на одну порцию (пример данных представлен ниже).
Напишите программу, которая будет запрашивать у пользователя количество порций для приготовления этих блюд и отображать информацию о суммарном количестве требуемых ингредиентов в указанном виде.

```python
cook_book = { 'салат': [ {'ingridient_name': 'сыр', 'quantity': 50, 'measure': 'гр'}, {'ingridient_name': 'томаты', 'quantity': 2, 'measure': 'шт'}, {'ingridient_name': 'огурцы', 'quantity': 20, 'measure': 'гр'}, {'ingridient_name': 'маслины', 'quantity': 10, 'measure': 'гр'}, {'ingridient_name': 'оливковое масло', 'quantity': 20, 'measure': 'мл'}, {'ingridient_name': 'салат', 'quantity': 10, 'measure': 'гр'}, {'ingridient_name': 'перец', 'quantity': 20, 'measure': 'гр'} ], 'пицца': [ {'ingridient_name': 'сыр', 'quantity': 20, 'measure': 'гр'}, {'ingridient_name': 'колбаса', 'quantity': 30, 'measure': 'гр'}, {'ingridient_name': 'бекон', 'quantity': 30, 'measure': 'гр'}, {'ingridient_name': 'оливки', 'quantity': 10, 'measure': 'гр'}, {'ingridient_name': 'томаты', 'quantity': 20, 'measure': 'гр'}, {'ingridient_name': 'тесто', 'quantity': 100, 'measure': 'гр'}, ], 'лимонад': [ {'ingridient_name': 'лимон', 'quantity': 1, 'measure': 'шт'}, {'ingridient_name': 'вода', 'quantity': 200, 'measure': 'мл'}, {'ingridient_name': 'сахар', 'quantity': 10, 'measure': 'гр'}, {'ingridient_name': 'лайм', 'quantity': 20, 'measure': 'гр'}, ] }

cnt_plates = int(input('Введите количество порций '))

# в данный словарь будем записывать наименования ингридиентов и граммовку с учетом количества порций
new_dic = {}

# первым циклом перебираем ингридиенты каждого блюда
for dish_ing in cook_book.values():
# вторым циклом перебираем ключи-значения по каждому ингридиенту 
  for ing_val in dish_ing:
# если ингиридент с данной размерностью есть в словаре, увеличиваем количество. Если инридиента с данной размерностью нет - добавляем в словарь 
      if (ing_val['ingridient_name'], ing_val['measure']) in new_dic:
        new_dic[(ing_val['ingridient_name'], ing_val['measure'])] = new_dic[(ing_val['ingridient_name'], ing_val['measure'])] + ing_val['quantity']*cnt_plates
      else:
        new_dic[(ing_val['ingridient_name'], ing_val['measure'])] = ing_val['quantity']*cnt_plates
# выводим результат - наименование инридиента, количество, размерность
for key, value in new_dic.items():
  print(key[0].capitalize(), value, key[1])
    
```
