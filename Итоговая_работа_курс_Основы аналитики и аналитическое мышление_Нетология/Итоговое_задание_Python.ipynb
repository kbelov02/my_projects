Получаем таблицу с названием города, кодом страны и населением, сразу отфильтруем по коду интересующих нас стран - в разных странах могут быть города с одинаковым названием. 



```python
df_c = pd.read_sql(
    """
    SELECT name,
           country_code,
           population
    FROM city
    WHERE country_code IN ('ARG',
                            'RUS',
                            'TUR',
                            'ESP',
                            'YUG',
                            'THA'
                             )

    """,
    "postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
df_c.head()
```

```python
#Загружаем датафрейм с ответами по отпускам. В шапке изменил название колонок на англ, чтобы было удобно джойнить
df = pd.read_csv('vac_plan.csv', sep=';')
df
```

```python
# Соединяем датафрейм с городами и с ответами по полям названия города
df_merged = df.merge(df_c, left_on='city_eng', right_on='name', how='inner')
df_merged.head(20)
```

```python
#Выводим новую таблицу с названием страны, городом и численностью населения
df_сoun_cit = df_merged[['country', 'city_rus', 'population']]
df_сoun_cit
```

```python
#Сочи задублировался, т.к. 2 раза в нем отдыхали. Надо удалить, иначе среднее посчитается неверно. 
df_uniq = df_сoun_cit.drop_duplicates(keep='first')
df_uniq
```

```python
#А теперь группируем по стране и считаем среднее
df_total = df_uniq.groupby('country')['population'].mean()
df_total
```
