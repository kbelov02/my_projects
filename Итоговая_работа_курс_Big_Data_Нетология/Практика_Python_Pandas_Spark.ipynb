```python
df_merged['release date']=pd.to_datetime(df_merged['release date'], errors='coerce')
```

```python
df_for_modeling = df_merged[['movie id', 'rating', 'unknown', 'Action', 'Adventure', 'Animation', 'Children_s', 'Comedy', 'Crime',
                             'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
                             'Thriller', 'War', 'Western', ('rating', 'count'),    ('rating', 'sum')]]
df_for_modeling.head()
```

```python
df_for_modeling = df_for_modeling.rename(columns={('rating', 'count'): 'count', ('rating', 'sum'): 'sum'})
df_for_modeling.head()
```

# Задание №5
Сформируйте X_train, X_test, y_train, y_test

```python
X, y = df_for_modeling[['movie id', 'unknown', 'Action', 'Adventure', 'Animation', 'Children_s', 'Comedy', 'Crime',
                             'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
                             'Thriller', 'War', 'Western', 'count', 'sum']], df_for_modeling[['rating']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.head()
```

# Задание №11
В спарке получить 2 датафрейма с 5-ю самыми популярными и самыми непопулярными фильмами (по количеству оценок, либо по самой оценке - на Ваш выбор)

```python
movie_cnt_rating = data.groupBy('item id').count()
```

```python
movie_cnt_rating.show()
```

```python
movie_cnt_rating = movie_cnt_rating.join(item, 'item id', how='inner')
```

```python
movie_cnt_rating = movie_cnt_rating.select('movie title', 'count')
movie_cnt_rating.show()
```

```python
from pyspark.sql.functions import desc, asc, col
```

```python
#5 самых популярных фильмов по количеству оценок
max_qty = movie_cnt_rating.sort(col('count').desc())
max_qty.show(5)
```

```python
# 5 самых непопулярных фильмов по количеству оценок
min_qty = movie_cnt_rating.sort(col('count').asc())
min_qty.show(5)
```
